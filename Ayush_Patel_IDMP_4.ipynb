```python
from pymongo import MongoClient
import pandas as pd
```

```python
client = MongoClient()
```

```python
restaurants_db = client["restaurants"]
restaurants = restaurants_db["restaurants"]
```

```python
cursor = restaurants.find()
df = pd.DataFrame(list(cursor))
df
```

```python
#Q1. 1)
cursor = restaurants.find({"borough": {"$in": ["Brooklyn","Bronx"]}}, {"_id":0, "name":1})

for i in cursor:
    print(i)
```

```python
#Q1. 2)
cursor = restaurants.find({"name":{"$regex" : "^Mad"}}, {"name": 1, "borough": 1, "cuisine": 1})

for i in cursor:
    print(i)
```

```python
#Q1. 3)
cursor = restaurants.find({"grades":{"$elemMatch": {"score":{"$gte":80,"$lte":90 }}}}, {"_id":0, "name":1})

for i in cursor:
    print(i)
```

```python
#Q1. 4)
import datetime
start = datetime.datetime(2014, 1, 1, 0, 0, 0, 0)
end = datetime.datetime(2015, 1, 1, 0, 0, 0, 0)
cursor = restaurants.find({"grades":{"$elemMatch":{"grade":"C","date":{"$gte":start,"$lt":end}}}},{"name":1,"_id":1})

for i in cursor:
    print(i)
    
```

```python
#Q1. 5)
cursor = restaurants.aggregate([{'$project':{'cuisine':{'$split':['$cuisine',',' '']},'name':1,'_id':0}},
{'$unwind': '$cuisine'}, {'$group':{'_id':'$cuisine','count':{'$sum':1}}}])

for i in cursor:
    print(i)
```

```python
#Q1. 6)
cursor = restaurants.aggregate([{'$match':{'cuisine':{'$ne':'American'}}},
{'$unwind':'$grades'}, {'$group':{'_id':'$restaurant_id','avg_score':{'$avg':'$grades.score'}}},
{'$match':{'avg_score':{'$gt':30}}}, {'$sort':{'avg_score':-1}}])

for i in cursor:
    print(i)
```

```python
#Q1. 7)
start = datetime.datetime(2014, 1, 1, 0, 0, 0, 0)
cursor = restaurants.aggregate([
{'$project':{'name':1,'grades':
{'$filter': {'input': '$grades','as': 'grade','cond':{'$gte': [ '$$grade.date', start ] }}}}},
{'$project':{ 'name':1,'_id':0,'Grades 2014 onwards':'$grades.grade'}}
]) 

for i in cursor:
    print(i)
```

```python
#Q1. 8)
cursor = restaurants.aggregate([{'$unwind':'$grades'},
{'$group':{'_id':'null','avg_score':{'$avg':'$grades.score'}}},
{'$project':{'_id':0}}])

for i in cursor:
    print(i)
```

```python
# Q3 For VIETNAM (SOUTH)
query = """
 SELECT ContryFlyingMission, MissionDate 
 FROM Bombing_Operations
 WHERE ContryFlyingMission = 'VIETNAM (SOUTH)'
"""
```

```python
#Q4 
query_takeoffs = """
 SELECT TakeOffLocation, MissionDate, TargetCountry 
 FROM Bombing_Operations 
 WHERE MissionDate = '1966-06-29'
 AND TargetCountry = 'NORTH VIETNAM'
"""
takeoffs = spark.sql(query_takeoffs) 
takeoffs = takeoffs.groupBy('TakeOffLocation').agg({'TargetCountry': 'count'}) 
takeoffs.show() 
```

```python
#Q5
mission_month = Bombing_Operations.withColumn('MissionMonth', Bombing_Operations.MissionDate[0:7]) 
mission_month.createOrReplaceTempView('missions') 
query_missions = """
 SELECT MissionMonth, COUNT(MissionMonth) FROM missions
 GROUP BY MissionMonth
 ORDER BY count(MissionMonth) DESC
"""
missions = spark.sql(query_missions) 
missions.show()
```

```python
#Q6
df = Bombing_Operations.groupby(Bombing_Operations["ContryFlyingMission"],Bombing_Operations["TakeoffLocation"],
                       Bombing_Operations["TargetCountry"],Bombing_Operations["MissionDate"]).sum("WeaponsLoadedWeight")
df = df.orderBy(df["sum(WeaponsLoadedWeight)"].desc())
df.show()
```
